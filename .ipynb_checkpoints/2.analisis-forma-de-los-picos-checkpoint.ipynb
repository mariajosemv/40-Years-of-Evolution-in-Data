En el notebook anterior se determinó que el tamaño del pico Geospiza *scandens* cambió en el transcurso de 37 años. Para explicar este cambio se tienen algunas hipótesis, entre ellas, se considera la sequía de 1976 y 1977 que provocó la muerte de algunas plantas que producen pequeñas semillas pequeñas en la isla. Las semillas grandes requerían picos de mayor tamaño para romperlas, por lo que los pájaros que tenian estas características sobrevivieron y luego se reprodujeron. Si este es el caso, es lógico pensar que la longitud del pico también cambie con el tiempo. De esto se desprenden otras dos hipótesis importantes:

- Si la longitud y el tamaño del pico cambian a la misma tasa, el pico tiene la misma forma; simplemente se hace más grande.

- Si la longitud y el tamaño del pico cambian a distintas tasas, la forma del pico cambia.

A continuación, se mostrará a través de un análisis exploratorio inicial como se relacionan la longitud y tamaño del pico en 1975 y 2012.

```python
import numpy as np # 
import pandas as pd # 
import matplotlib.pyplot as plt 
import seaborn as sns 

np.random.seed(42)

%matplotlib inline

sns.set() 

# Importar las funciones matematicas definidas para el analisis
%run funciones-matematicas.ipynb

# Importar datos
data = pd.read_csv('./datasets/data_finch_beaks.csv')

# Ajustar tamaño de las figuras en matplotlib
plt.rcParams['figure.figsize'] = [10, 8]
```

## Análisis exploratorio

Relación entre la longitud y el tamaño del pico en 1975 y 2012.

```python
# Seleccionar la especie Geospiza scandens
scandens = data[data['species']== 'scandens']

# Seleccionar el tamaño de los picos en 1975 y 2012
bd_1975 = scandens[scandens['year'] == 1975]['Beak depth, mm']
bd_2012 = scandens[scandens['year'] == 2012]['Beak depth, mm']

# Seleccionar la longitud de los picos en 1975 y 2012
bl_1975 = scandens[scandens['year'] == 1975]['Beak length, mm']
bl_2012 = scandens[scandens['year'] == 2012]['Beak length, mm']

# Scatter plot de los datos de 1975 
_ = plt.plot(bl_1975, bd_1975, marker='.', linestyle='None', alpha=0.5, color='blue')

# Scatter plot de los datos de 2012 
_ = plt.plot(bl_2012, bd_2012, marker='.', linestyle='None', alpha=0.5, color='red')

# Ajustes de ejes y leyenda
_ = plt.xlabel('Longitud (mm)')
_ = plt.ylabel('Tamaño (mm)')
_ = plt.legend(('1975', '2012'), loc='upper left')


plt.show()
```

En la grafica se observa que algunos puntos rojos se ubican más arriba en el eje `Y`, alrededor de los 13-14 mm de longitud; esto quiere decir que el tamaño de los picos ha aumentado, pero no tanto así su longitud si se compara con los puntos azules. Por otra parte, se tienen otras muestras que han disminuido un poco su longitud debido a que se ubican a la izquierda de los puntos azules, en los alrededores de los 12 y 13 mm en el eje `X`. En conclusión podemos ver que los picos **no han mantenido la misma forma** que en 1975. 

## Regresión lineal

Para cuantificar y establecer la diferencia entre las relaciones tamaño/longitud de los picos, se ajustarán los datos a una regresión lineal con un intervalo de confianza del 95%. 

```python
# Calculo de la regresión lineal: pendiente y punto de corte
slope_1975, intercept_1975 = np.polyfit(bl_1975, bd_1975, 1)
slope_2012, intercept_2012 = np.polyfit(bl_2012, bd_2012, 1)

# Cálculo de los bootstrap replicates
bs_slope_reps_1975, bs_intercept_reps_1975 = draw_bs_pairs_linreg(bl_1975, bd_1975, size=1000)
bs_slope_reps_2012, bs_intercept_reps_2012= draw_bs_pairs_linreg(bl_2012, bd_2012, size=1000)

# Cálculo del intervalo de confianza de la pendiente y punto de corte
slope_conf_int_1975 = np.percentile(bs_slope_reps_1975, [2.5, 97.5])
slope_conf_int_2012 = np.percentile(bs_slope_reps_2012, [2.5, 97.5])

intercept_conf_int_1975 = np.percentile(bs_intercept_reps_1975, [2.5, 97.5])
intercept_conf_int_2012 = np.percentile(bs_intercept_reps_2012, [2.5, 97.5])

# Scatter plot de los datos de 1975 
_ = plt.plot(bl_1975, bd_1975, marker='.',
             linestyle='none', color='blue', alpha=0.5)

# Scatter plot de los datos de 2012
_ = plt.plot(bl_2012, bd_2012, marker='.',
             linestyle='none', color='red', alpha=0.5)

# Ajustes de ejes y leyenda
_ = plt.xlabel('Longitud (mm)')
_ = plt.ylabel('Tamaño (mm)')
_ = plt.legend(('1975', '2012'), loc='upper left')

# Líneas para los bootstrap replicates
x = np.array([10, 17])

# Visualización de la simulación
for i in range(100):
    plt.plot(x, bs_slope_reps_1975[i]*x + bs_intercept_reps_1975[i],
             linewidth=0.5, alpha=0.2, color='blue')
    plt.plot(x, bs_slope_reps_2012[i]*x + bs_intercept_reps_2012[i],
             linewidth=0.5, alpha=0.2, color='red')

plt.show()



# Mostrar resultados
print('1975: pendiente:', slope_1975,
      'intervalo de confianza:', slope_conf_int_1975)
print('1975: punto de corte:', intercept_1975,
      'intervalo de confianza:', intercept_conf_int_1975)
print()
print('2012: pendiente:', slope_2012,
      'intervalo de confianza:', slope_conf_int_2012)
print('2012: punto de corte:', intercept_2012,
      'intervalo de confianza:', intercept_conf_int_2012)
```

Los resultados muestran que las relaciones tienen pendientes similares (0.46 mm), pero difieren en el punto de corte. Esto sugiere que el aumento en el tamaño de los picos es proporcional al aumento de la longitud. En otras palabras, por cada milimetro de aumento en la longitud del pico, se tiene aproximadamente 0.5 mm de aumento en el tamaño de los picos para ambos años.

Finalmente, para analizar el cambio en la forma del pico, comparamos el ratio de la longitud y el tamaño en 1975 y 2012 con un 99% del nivel de confianza.

```python
# Cálculo de ratio de longitud/tamaño
ratio_1975 = bl_1975 / bd_1975
ratio_2012 = bl_2012 / bd_2012

# Cálculo de promedios
mean_ratio_1975 = np.mean(ratio_1975)
mean_ratio_2012 = np.mean(ratio_2012)

# Bootstrap replicates para el promedio
bs_replicates_1975 = draw_bs_reps(ratio_1975, np.mean, size=10000)
bs_replicates_2012 = draw_bs_reps(ratio_2012, np.mean, size=10000)

# Cálculo del intervalo de confianza al 99%
conf_int_1975 = np.percentile(bs_replicates_1975, [0.5, 99.5])
conf_int_2012 = np.percentile(bs_replicates_2012, [0.5, 99.5])

# Visualización de los intervalos de confianza
_ = plt.plot((conf_int_1975[0], conf_int_1975[1]), (1975,1975), color = 'r', linewidth=3)
_ = plt.plot((conf_int_2012[0], conf_int_2012[1]), (2012,2012), color = 'b', linewidth=3)

# Scatter plot ratio longitud/tamaño del pico
_ = plt.plot(mean_ratio_1975, 1975, marker='o', color = 'r')
_ = plt.plot(mean_ratio_2012, 2012, marker='o', color = 'b')

# Ajustes de ejes
_ = plt.margins(x=0.05, y=0.2)
_ = plt.yticks([1975, 2012])
_ = plt.xlabel('Longitud/Tamaño del pico')


# Mostrar Resultados
print('1975: mean ratio =', mean_ratio_1975,
      'conf int =', conf_int_1975)
print('2012: mean ratio =', mean_ratio_2012,
      'conf int =', conf_int_2012)
```

Estos resultados muestran que el ratio promedio de la longitud y tamaño del pico disminuyo aproximadamente 0,11 mm o un 7%, de 1975 a 2012. Por otra parte, los intervalos de confianza del 99% no están ni siquiera cerca de superponerse, por lo que se trata de un cambio real. **La forma del pico ha cambiado**: han disminuido en longitud y aumentado su tamaño.
